Deskripsi Kasus
    Dataset ini dikumpulkan melalui survei yang bertujuan memahami hubungan antara media sosial dengan kesehatan mental. Survei ini terdiri dari beberapa variabel yang mengukur berbagai aspek, seperti tingkat depresi, konsentrasi, dan lainnya. Berikut penjelasan mengenai beberapa variabel penting dalam dataset ini: 
    
1. Timestamp: Waktu pengisian survei (format tanggal dan waktu)

2. Age: Usia responden (angka kontinu)

3. Gender: Jenis kelamin responden (kategorikal)

4. Relationship: Status hubungan responden (kategorikal)

5. Occupation: Pekerjaan responden (kategorikal)

6. Affiliation: Afiliasi responden (misalnya, Universitas, Perusahaan, Sekolah, dll.) (kategorikal)
    
7. SocialMedia: Apakah responden menggunakan media sosial? (biner: Ya/Tidak)

8. Platforms: Platform media sosial yang digunakan responden (kategorikal)

9. TimeSpent: Durasi waktu yang dihabiskan responden di media sosial (kategorikal: "Less than an Hour", "Between 1 and 2 hours", "Between 2 and 3 hours", "Between 3 and 4 hours", "Between 4 and 5 hours", "More than 5 hours")

10. NoPurposeUse: Tingkat penggunaan media sosial tanpa tujuan (angka kontinu, skala 1-5)

11. Distraction: Tingkat gangguan yang dirasakan dari media sosial (angka kontinu, skala 1-5)

12. Restlessness: Tingkat kegelisahan yang dirasakan dari media sosial (angka kontinu, skala 1-5)

13. EasilyDistracted: Tingkat mudah terdistraksi yang dirasakan dari media sosial (angka kontinu, skala 1-5)

14. Worries: Tingkat kekhawatiran yang dirasakan dari media sosial (angka kontinu, skala 1-5)

15. Concentration: Tingkat gangguan konsentrasi yang dirasakan dari media sosial (angka kontinu, skala 1-5)

16. Comparison: Tingkat perbandingan diri dengan orang lain di media sosial (angka kontinu, skala 1-5)

17. ComparisonFeel: Perasaan yang muncul setelah membandingkan diri dengan orang lain di media sosial (angka kontinu, skala 1-5)

18. Validation: Tingkat validasi yang dicari di media sosial (angka kontinu, skala 1-5)

19. Depression: Tingkat depresi yang dirasakan dari media sosial (angka kontinu, skala 1-5)

20. InterestFluctuation: Tingkat fluktuasi minat yang dirasakan dari media sosial (angka kontinu, skala 1-5)

21. SleepIssues: Tingkat masalah tidur yang dirasakan dari media sosial (angka kontinu, skala 1-5)

*impor library yang akan digunakan*

```python

import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

*Mengimpor data yang akan digunakan*

```python

data_survey = pd.read_csv('D:/COOLYEAH/Semester 3/data_survey.csv')
data_survey.head()
```

*Preprocessing Data*

```python
# mengecek informasi mengenai data
data_survey.info()
```

```python
# melihat apakah ada missing value dalam data
np.sum(data_survey.isnull())
```

```python
# mengatasi missing value pada kolom Affiliation
data_survey["Affiliation"]. fillna("None", inplace=True)
```

```python
#mengecek lagi apakah masih ada misssing value
np.sum(data_survey.isnull())
```

```python
# menghapus kolom yang tidak perlu
data_survey.drop(columns = ['Unnamed: 0'], inplace=True)
data_survey.drop(columns = ['Timestamp'], inplace=True)
data_survey.drop(columns = ['Relationship'], inplace=True)
```

```python
# mengecek apakah kolom sudah terhapus
data_survey.head()
```

```python
# menghapus data responden yang tidak menggunakan sosial media 
data_survey = data_survey[data_survey['SocialMedia'] !='No']
```

```python
#cek apakah baris sudah terhapus
data_survey.info()
```

```python
#Memisahkan jenis platform dalam 1 sel
list_platform = []
for platform in data_survey['Platforms']:
    list_platform.extend(platform.split(', '))
```

```python
#Menghitung jumlah jenis kelamin kolom Gender
#hitung_gender = pd.Series('Gender').value_counts()
#hitung_gender

data_survey['Gender'].unique()
```

```python
#Menghitung jumlah jenis platform kolom list_platform
hitung_platform = pd.Series(list_platform).value_counts()
hitung_platform
```

```python
#mengganti nama kategori dalam kolom timespent 
time_spent = {
    "Less than an Hour": "<1 hour",
    "Between 1 and 2 hours": "1-2 hour",
    "Between 2 and 3 hours": "2-3 hour",
    "Between 3 and 4 hours": "3-4 hour",
    "Between 4 and 5 hours": "4-5 hour",
    "More than 5 hours": ">5 hour"
}

# membuat kolom baru untuk menyimpan kolom yang baru dari kolom timespent
data_survey['TimeSpentCategory'] = data_survey['TimeSpent'].replace(time_spent)
data_survey.head()
```

```python
# menghapus kolom timespent
data_survey.drop(columns = ['TimeSpent'], inplace=True)
data_survey.head()
```

```python
# buat urutan untuk TimeSpentCategory
time_order = ['<1 hour', '1-2 hour', '2-3 hour', '3-4 hour', '4-5 hour', '>5 hour']
```

1. membuat pie chart untuk melihat presentase pekerjaan responden

```python
# mengetahui apa saja jenis pekerjaan responden dalam kolom Occupation
hitung_pekerjaan = data_survey['Occupation'].value_counts()
hitung_pekerjaan
```

```python
# Buat pie chart
plt.pie(
    hitung_pekerjaan.values,  # nilai untuk setiap slice
    labels=hitung_pekerjaan.index,  # label untuk setiap slice
    autopct='%1.1f%%',  # menampilkan persentase
)

plt.title('Presentase Pekerjaan Responden')  # membuat judul
plt.axis('equal')  # membuat pie chart berbentuk lingkaran

plt.show()
```

Dari diagram diatas dapat diketahui jika, presentase pekerjaan responden terbagi menjadi 4 kategori yaitu :
    - Mahasiswa: 60.9% dari responden adalah mahasiswa
    - Pekerja Kantoran: 27.2% dari responden adalah pekerja kantoran
    - Pelajar: 10.3% dari responden adalah pelajar
    - Pensiunan: 1.7% dari responden adalah pensiunan
sehingga, dapat kita simpulkan jika mayoritas responden merupakan mahasiswa

2. membuat pie chart untuk mengetahui apakah waktu menggunakan media sosial berhubungan dengan kesehatan mental

```python
# membuat Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='TimeSpentCategory', y='Depression', data=data_survey, order=time_order, ci=None)

plt.title('Rata-rata tingkat depresi berdasarkan waktu yang dihabiskan di media sosial')
plt.xlabel('Waktu di Media Sosial')
plt.ylabel('Rata-rata Tingkat Depresi')
plt.xticks(rotation=45)

plt.show()
```

Dari diagram bar ini, menunjukkan bahwa durasi waktu bermain media sosial memengaruhi tingkat depresi seseorang. Dapat kita lihat, jika semakin banyak waktu yang dihabiskan, maka semakin tinggi juga tingkat depresinya

4. Apakah pekerjaan atau occupation dapat memengaruhi tingkat validasi dengan memperhatikan jenis kelamin

```python
# memfilter gender yang digunakan cuma perempuan dan laki-laki saja
filter_gender = data_survey[data_survey['Gender'].isin(['Female', 'Male'])]

# mengelompokkan data berdasarkan 'Occupation'
grouped_data = filter_gender.groupby('Occupation')['Comparison'].mean().reset_index()

# membuat diagram
plt.figure(figsize=(10, 6))
sns.barplot(x='Occupation', 
            y='Comparison', 
            hue='Gender', 
            data=filter_gender, 
            ci=None,  
            palette={'Female': '#FF9874', 'Male': '#7695FF'})


plt.title("Rata-rata tingkat validasi berdasarkan pekerjaan dan gender")
plt.xlabel('Pekerjaan')
plt.ylabel('Tingkat Validasi')
plt.xticks(rotation=45)

plt.show()
```

Dari diagram diatas dapat kita ketahui jika pekerjaan dan jenis kelamin seseorang memengaruhi tingkat validasi yang dicari di media sosial. Mayoritas responden dengan jenis kelamin perempuan cenderung mencari validasi yang tinggi di media sosial. Terlihat juga bahwa mahasiswa memiliki tingkat validasi yang relatif lebih tinggi dibandingkan dengan pelajar, pekerja kantoran, dan pensiunan

7. Mengetahui perbandingan tingkat gangguan konsentrasi berdasarkan jenis kelamin, sesuai dengan pekerjaan

```python
# Mengelompokkan data berdasarkan 'Occupation' dan 'Gender'
grouped_data = filter_gender.groupby(['Occupation', 'Gender'])['Concentration'].mean().reset_index()

# Membuat diagram line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Occupation', 
             y='Concentration', 
             hue='Gender', 
             data=grouped_data, 
             marker='o', 
             palette={'Female': '#FF9874', 'Male': '#7695FF'})

# Menambahkan judul dan label pada chart
plt.title("Rata-rata tingkat gangguan konsentrasi berdasarkan pekerjaan dan gender")
plt.xlabel('Pekerjaan')
plt.ylabel('Tingkat Konsentrasi')
plt.xticks(rotation=45)
plt.show()
```

Grafik garis ini menunjukkan rata-rata tingkat gangguan konsentrasi berdasarkan pekerjaan dan gender.Secara umum, perempuan mengalami gangguan konsentrasi yang lebih tinggi dibandingkan dengan laki-laki di semua kategori pekerjaan.Rata-rata gangguan konsentrasi tertinggi dialami oleh perempuan pensiunan, disusul perempuan pekerja kantoran, perempuan pelajar, dan terakhir perempuan mahasiswa.Sementara itu, laki-laki mengalami gangguan konsentrasi yang lebih tinggi pada kategori pekerja kantoran, kemudian mahasiswa, pelajar, dan terendah pada pensiunan
